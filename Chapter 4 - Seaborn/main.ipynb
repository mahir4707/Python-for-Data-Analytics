```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
#scatter plot for age vs fare
sns.set_style("ticks") #used to style the background of graph - white, whitegrid, dark, darkgrid, ticks

sns.set_context("notebook") #terms notebook, paper, talk, poster
sns.scatterplot(data = data, x="age", y="fare", hue="alive", palette="Dark2") #pallete - used for color choice - set1, Set2, Dark2, 
sns.despine() #used to remove the left or right boundry - (position = true)
```

```python
sns.scatterplot(data = data, x="age", y="fare", hue="alive",style="embark_town") 
```

```python
# Strip Plot:
# Used to visualize individual data points for a numerical variable across one or more categorical variables.
# It shows the raw distribution of data and helps identify spread, clustering, and potential outliers.
# Best suited for categorical vs numerical analysis during Exploratory Data Analysis (EDA).
# Points may overlap when the dataset is large.
#
# General Syntax:
# sns.stripplot(x=<categorical_column>, y=<numerical_column>, hue=<categorical_column>, data=<dataframe>)

sns.stripplot(data = data, x="fare", y="embark_town", hue="alive", jitter=True, dodge=True)
```

```python
# Swarm Plot:
# Similar to strip plot but automatically adjusts the position of points to avoid overlap.
# Provides a clearer view of the distribution and density of data points within each category.
# Ideal for small to medium-sized datasets where showing every observation matters.
#
# General Syntax:
# sns.swarmplot(x=<categorical_column>, y=<numerical_column>, hue=<categorical_column>, data=<dataframe>)

sns.swarmplot(data=data, x="survived", y="fare", hue="sex" )
```

```python
#line plot
# data.head()
sns.set_style("whitegrid")
sns.lineplot(data=data, x="age", y="fare", hue="sex")
```

```python
#Histogram - gives the distribution of data along with our x
#here histogram gives the distribution of data for age

sns.histplot(data=data, x="age", hue="alive", multiple="stack")
```

```python
#regression plot - 

sns.regplot(data=data, x="age", y="fare")
```

```python
#Joint Plot
sns.jointplot(data =data, x="age", y="fare", hue="sex")
```
