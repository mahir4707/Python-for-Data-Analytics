```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
data["class"].value_counts()
```

```python
data["alive"].value_counts()
```

```python
data["age"].value_counts()
```

```python
#scatter plot for age vs fare
sns.set_style("ticks") #used to style the background of graph - white, whitegrid, dark, darkgrid, ticks

sns.set_context("notebook") #terms notebook, paper, talk, poster
sns.scatterplot(data = data, x="age", y="fare", hue="alive", palette="Dark2") #pallete - used for color choice - set1, Set2, Dark2, 
sns.despine() #used to remove the left or right boundry - (position = true)
```

```python
sns.scatterplot(data = data, x="age", y="fare", hue="alive",style="embark_town") 
```

```python
# Strip Plot:
# Used to visualize individual data points for a numerical variable across one or more categorical variables.
# It shows the raw distribution of data and helps identify spread, clustering, and potential outliers.
# Best suited for categorical vs numerical analysis during Exploratory Data Analysis (EDA).
# Points may overlap when the dataset is large.
#
# General Syntax:
# sns.stripplot(x=<categorical_column>, y=<numerical_column>, hue=<categorical_column>, data=<dataframe>)

sns.stripplot(data = data, x="fare", y="embark_town", hue="alive", jitter=True, dodge=True)
```

```python
# Swarm Plot:
# Similar to strip plot but automatically adjusts the position of points to avoid overlap.
# Provides a clearer view of the distribution and density of data points within each category.
# Ideal for small to medium-sized datasets where showing every observation matters.
#
# General Syntax:
# sns.swarmplot(x=<categorical_column>, y=<numerical_column>, hue=<categorical_column>, data=<dataframe>)

sns.swarmplot(data=data, x="survived", y="fare", hue="sex" )
```

```python
#line plot
# data.head()
sns.set_style("whitegrid")
sns.lineplot(data=data, x="age", y="fare", hue="sex")
```

```python
#Histogram - gives the distribution of data along with our x
#here histogram gives the distribution of data for age

sns.histplot(data=data, x="age", hue="alive", multiple="stack")
```

```python
#regression plot - 

sns.regplot(data=data, x="age", y="fare")
```

```python
#Joint Plot
sns.jointplot(data =data, x="age", y="fare", hue="sex")
```

```python
#Barplot
import numpy as np
sns.barplot(data=data, x="class", y="age", hue="sex", estimator=np.mean, palette=["skyblue", "pink"])
```

```python
#countplot

sns.countplot(data=data, x="alive", hue="sex")
```

```python
#box plot - used to find out the outliers in the data

sns.boxplot(data=data, x="class", y="age", hue="sex")
```

```python
#Violin plot 

sns.violinplot(data=data, x="class", y="age", hue="sex", palette="pastel", split=True)
sns.swarmplot(data=data, x="class", y="age", hue="sex", palette="dark", size=3)
sns.move_legend(plt.gca(), "upper left", bbox_to_anchor=(0.6, 1))
```

```python
#kde plot - Kernel Density Estimation Plot
#Description:
#A KDE plot is a smoothed version of a histogram that estimates the probability density function of a continuous variable.
#It provides insights into the distribution and density of data points.
#General Syntax:
#sns.kdeplot(x=<numerical_column>, hue=<categorical_column>, data=<dataframe>)

sns.kdeplot(data=data, x="age", hue="alive", fill=True)
```

```python



columns = ["age", "fare", "survived", "alone"]

data[columns].corr()
```

```python
#Heatmap - used to find correlation between different numerical variables

columns = ["age", "fare", "survived", "alone"]

sns.heatmap(data[columns].corr(), annot=True, cmap="YlGnBu", linewidths=1, linecolor="black")
```

```python
#rug plot  
#description:
#A rug plot is a one-dimensional representation of data points along an axis, typically used to visualize the distribution of a single variable.
#It displays individual data points as small vertical lines (or "rugs") along the x-axis, providing insights into the density and distribution of the data.
# General Syntax:
# sns.rugplot(x=<numerical_column>, hue=<categorical_column>, data=<dataframe>)

sns.rugplot(data=data, x="age",hue="sex", height=0.2)
```

```python
#Pair plot 
#Description:
#A pair plot is a grid of scatter plots that visualizes pairwise relationships between multiple numerical variables in a dataset.
#It helps identify correlations, distributions, and potential patterns among the variables.     
columns = ["age", "fare", "survived", "alone"]
sns.pairplot(data=data[columns], hue="survived", height=2.5, diag_kind="kde")
```

```python
#pair grid 
#description:
#A PairGrid is a flexible grid layout for creating multiple plots that visualize pairwise relationships between variables in a dataset.
#It allows for customized plotting of different types of relationships (e.g., scatter plots, histograms) across multiple variables, facilitating comprehensive exploratory data analysis.
# General Syntax:       
columns = ["age", "fare", "survived", "alone"]
pair_grid = sns.PairGrid(data=data[columns], hue="survived", height=2.5)
pair_grid.map_upper(sns.scatterplot)
pair_grid.map_lower(sns.kdeplot)
pair_grid.map_diag(sns.histplot)
pair_grid.add_legend()
```
