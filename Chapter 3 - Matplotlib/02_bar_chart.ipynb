```python
import matplotlib.pyplot as plt
```

```python
# 1. Number of employees in each department (Bar Chart)
departments = ["IT", "HR", "Finance", "Marketing", "Operations"]
employee_count = [18, 6, 9, 7, 10]

plt.bar(departments, employee_count)
plt.title("Number of Employees in each department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")


plt.show()
```

```python
# 2. Total sales for different product categories (Bar Chart)
product_categories = ["Furniture", "Office Supplies", "Technology"]
total_sales = [250000, 180000, 320000]

plt.bar(product_categories, total_sales)
plt.title("Category wise sales")
plt.xlabel("Product Category")
plt.ylabel("Sales")
plt.show()
```

```python
# 1. Top 5 countries by population (Horizontal Bar Chart)
countries = ["China", "India", "United States", "Indonesia", "Pakistan"]
population = [1412000000, 1408000000, 331000000, 276000000, 241000000]

plt.barh(countries, population)
plt.title("Top 5 countries by population")
plt.xlabel("Population")
plt.ylabel("Country Name")

plt.show()
```

```python
# 2. Skills count in a company (Horizontal Bar Chart)
skills = ["Python", "SQL", "Excel", "Power BI", "Machine Learning"]
skill_count = [18, 15, 22, 10, 8]

plt.barh(skills, skill_count)
plt.title("Skills count in company")
plt.xlabel("Skills count")
plt.ylabel("Skill name")

plt.show()
```
