```python
import matplotlib.pyplot as plt
```

```python
# 1. Salary distribution of employees (Histogram)
employee_salaries = [
    28000, 32000, 35000, 38000, 42000, 45000,
    48000, 52000, 58000, 60000, 65000, 70000,
    78000, 82000, 90000
]

plt.hist(employee_salaries)
plt.title("Salary Distribution of employees")
plt.show()
```

```python
# 2. Marks distribution of students (Histogram)
student_marks = [
    35, 42, 48, 55, 60, 65, 68, 72,
    75, 78, 82, 85, 88, 90, 95
]

plt.hist(student_marks)
plt.title("Marks Distribution of students")
plt.show()
```
