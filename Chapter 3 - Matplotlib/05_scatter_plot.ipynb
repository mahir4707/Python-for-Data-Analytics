```python
import matplotlib.pyplot as plt
```

```python
# 1. Age vs Salary of Employees (Scatter Plot)
employee_age = [22, 25, 28, 30, 32, 35, 38, 40, 45]
employee_salary = [28000, 35000, 42000, 50000, 58000, 68000, 75000, 82000, 95000]

plt.scatter(employee_age, 
            employee_salary, 
            marker="o", 
            alpha=0.8, 
            s=50,
            c='blue', 
            edgecolors='black',
            label='Age vs Salary' )
plt.title("Age vs Salary of Employees")
plt.xlabel("Age of Employee")
plt.ylabel("Salary of Employee")
plt.legend()
plt.show()
```

```python
# 2. Study Hours vs Exam Score (Scatter Plot)
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_score = [45, 52, 60, 68, 75, 82, 88, 93]

plt.scatter(study_hours,
            exam_score, 
            s=50,
            marker="s",
            c='r'
            )
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")

plt.show()
```
