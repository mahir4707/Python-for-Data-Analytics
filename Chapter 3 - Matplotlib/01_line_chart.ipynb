```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
monthly_sales = [12000, 13500, 12800, 15000, 16500, 17200,
                 16000, 18000, 19000, 21000, 22000, 24000]

plt.plot(months, monthly_sales, color='blue', linestyle='--', linewidth=2, marker='o')
plt.title("Sales over Year")
plt.xlabel("Month of the year")
plt.ylabel("Sales in thousands")
plt.show()
```

```python
# 2. Daily website visitors over a week (Line Chart)
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
website_visitors = [450, 520, 480, 610, 700, 850, 900]

plt.plot(days, website_visitors, marker="o", )
plt.title("Site visitor over a week")
plt.xlabel("Days of a week")
plt.ylabel("Website Visitor")
plt.grid()
plt.show()
```

```python
# 3. Temperature change over 10 days (Line Chart)
day_numbers = list(range(1, 11))
temperature = [22, 23, 24, 26, 27, 28, 29, 30, 28, 26]

plt.plot(day_numbers, temperature, color='red', marker="^")
plt.title("Change in Temp. past 10 days")
plt.xlabel("Day number")
plt.ylabel("Temperture")
plt.grid()
plt.show()
```

```python
# 1. Sales of Product A and Product B over the same months (Multiple Line Chart)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

product_a_sales = [12000, 13000, 12500, 14000, 15000, 15500,
                    14800, 16000, 17000, 18500, 19500, 21000]

product_b_sales = [10000, 11500, 11000, 12500, 13500, 14000,
                    13800, 15000, 15800, 17000, 18000, 19500]

plt.plot(months, product_a_sales, color="red", marker="o",  linestyle="--")
plt.plot(months, product_b_sales, color="blue", marker="o")
plt.title("Sales of Product A and B")
plt.xlabel("Month of the year")
plt.ylabel("Sales")


plt.show()
```

```python
# 2. Actual sales vs predicted sales over time (Multiple Line Chart)
time_period = ["Week 1", "Week 2", "Week 3", "Week 4", "Week 5", "Week 6"]

actual_sales = [5000, 5200, 5100, 5400, 5600, 5900]
predicted_sales = [4800, 5150, 5050, 5350, 5500, 5800]

plt.plot(time_period, actual_sales)
plt.plot(time_period, predicted_sales)
plt.title("Actual sales vs Predicted sales")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.grid()

plt.show()
```
