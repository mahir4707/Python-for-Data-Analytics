```python
import matplotlib.pyplot as plt
```

```python
# 1. Expense distribution of a company (Pie Chart)
expenses = ["Rent", "Salaries", "Marketing", "Utilities", "Miscellaneous"]
expense_amount = [30000, 50000, 15000, 8000, 7000]

plt.title("Expense distribution")
plt.pie(expense_amount, labels=expenses, colors=["red", "green", "blue", "yellow", "pink"], autopct='%1.1f%%')
plt.legend(loc='upper left',  bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# 2. Market share of different brands (Pie Chart)
brands = ["Brand A", "Brand B", "Brand C", "Brand D"]
market_share = [40, 30, 20, 10]

plt.pie(market_share, labels=brands, autopct='%1.1f%%')
plt.title("Market Share")
plt.legend(loc="upper right", bbox_to_anchor=(1.2,1))
plt.tight_layout()
plt.show()
```
